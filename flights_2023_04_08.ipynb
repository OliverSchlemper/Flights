```python
import pandas as pd
from FlightTracker import FlightTracker
from rnog_data.runtable import RunTable


tracker = FlightTracker('2022-04-01 00:00:00', '2022-08-01 00:00:00')
```

```python
tracker.flights_distinct
```

```python
runs = pd.DataFrame(['station', 'run'])
rnog_table = RunTable()
length = len(tracker.flights_distinct)
for i in range(length):
    mintime = tracker.flights_distinct.mintime.iloc[i]
    maxtime = tracker.flights_distinct.maxtime.iloc[i]
    df = pd.DataFrame(['station', 'run'])
    runtable = rnog_table.get_table(start_time=mintime, stop_time=maxtime)

    if len(runtable) != 0:
        df['station'] = runtable.station
        df['run'] = runtable.run
        runs = pd.concat([runs, df], ignore_index=True, sort=False)
    print(i, 'of', length)
runs
```

```python
runs = runs[['station', 'run']].copy()
runs.drop_duplicates(inplace = True)
runs.dropna(inplace = True)
```

```python
runs
```

```python
runs['station'] = runs.station.astype(int)
runs['run'] = runs.run.astype(int)
```

```python
runs = runs.sort_values(by = ['station', 'run'])
```
