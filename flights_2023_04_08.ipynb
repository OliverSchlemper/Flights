```python
import pandas as pd
from FlightTracker import FlightTracker
from rnog_data.runtable import RunTable


tracker = FlightTracker('2022-05-01 00:00:00', '2022-11-01 00:00:00')
```

```python
tracker.flights_distinct
```

```python
runs = pd.DataFrame({'station' : [], 'run' : []})
rnog_table = RunTable()
length = len(tracker.flights_distinct)
for i in range(length):
    mintime = tracker.flights_distinct.mintime.iloc[i]
    maxtime = tracker.flights_distinct.maxtime.iloc[i]
    df = pd.DataFrame()
    runtable = rnog_table.get_table(start_time=mintime, stop_time=maxtime)

    if len(runtable) != 0:
        df = pd.DataFrame({'station' : list(runtable.station), 'run' : list(runtable.run)})
        runs = pd.concat([runs, df], ignore_index=True, sort=False)
    print(i, 'of', length)
runs
```

```python
runs.drop_duplicates(inplace = True)
runs.dropna(inplace = True)

runs['station'] = runs.station.astype(int)
runs['run'] = runs.run.astype(int)
```

```python
runs = runs.sort_values(by = ['station', 'run'])
```

```python
runs
```

```python
import pandas as pd
runs_old = pd.read_csv('runs.csv', delimiter='\t')
runs_old.sort_values(by = ['station', 'run'])
runs_old.drop_duplicates(inplace=True)
runs_old
```

```python
combined = pd.concat([runs, runs_old], ignore_index=True)
# Drop those rows from df1

runs_filtered = combined.drop_duplicates(keep=False)
runs_filtered
```

```python
runs_filtered.to_csv('runs_filtered.csv', index=False)
```
